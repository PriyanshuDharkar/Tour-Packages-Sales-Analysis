```python
import pandas as pd
import sqlite3

# Read CSV files
Enquiry_details = pd.read_csv(r"C:\Users\ACER\Downloads\Sample Data(1).csv")

# Connect to SQLite database
conn = sqlite3.connect('relational_database.db')

# Write dataframes to SQL tables
Enquiry_details.to_sql('Enquiry_details', conn, index=False, if_exists='replace')


# Commit changes
conn.commit()
```

```python
def run_query(query):
    return pd.read_sql_query(query,conn)
```

```python
# Getting all the data
query = '''
SELECT *
FROM Enquiry_details
'''

run_query(query)
```

```python
# Agents with Number of leads
query = '''
SELECT Assigned_To AS Agent, COUNT(*) AS Total_leads
FROM Enquiry_details
GROUP BY 1
LIMIT 10
'''

run_query(query)
```

```python
# Top Agents with highest number of leads
query = '''
SELECT Assigned_To AS Agent, COUNT(*) AS Total_leads
FROM Enquiry_details
GROUP BY 1
ORDER BY 2 DESC
'''

run_query(query)

# Sales_Person 0 is getting most number of leads
```

```python
# Higest Conversion rate per agent 
query = '''
SELECT Assigned_To AS Agent, ROUND(SUM(CASE WHEN Status = 'Converted' THEN 1 ELSE 0 END)*100.0/COUNT(*),2) AS Converted_pct
FROM Enquiry_details
GROUP BY 1
ORDER BY 2 DESC
'''

run_query(query)

# But Sales Person 33 has the highest conversion rate whereas Sales Person 0 has an average conversion rate.
```

```python
# Revenue generated by trips
query = '''
SELECT Trip, COUNT(*) AS Total_Leads, 
SUM(CASE WHEN Status = 'Converted' THEN 1 ELSE 0 END) AS Total_Conversions,
SUM(Total_Bill) AS Revenue
FROM Enquiry_details
GROUP BY 1
ORDER BY 4 DESC
'''

run_query(query)
```

```python
# Revenue generated per Agent
query = '''
SELECT Assigned_To, SUM(Total_Bill) AS Revenue
FROM Enquiry_details
GROUP BY 1
ORDER BY 2 DESC
'''

run_query(query)
```

```python
# Monthly analysis of leads,conversion,RPL, ASP
query = '''
SELECT SUBSTR(Lead_Date,4,2) AS Month,
       COUNT(*) AS Total_Leads,
       SUM(CASE WHEN Status = 'Converted' THEN 1 ELSE 0 END) AS Total_Conversions,
       ROUND(SUM(CASE WHEN Status = 'Converted' THEN 1 ELSE 0 END)*100.0/COUNT(*),2) AS Conversion_pct,
       ROUND(SUM(Total_Bill),2) AS Total_Revenue, ROUND(SUM(Total_Bill)/ COUNT(*),2) AS Revenue_per_lead, 
       ROUND(SUM(Total_Bill)/SUM(CASE WHEN Status = 'Converted' THEN 1 ELSE 0 END),2) AS Avg_Selling_Price
FROM Enquiry_details
GROUP BY Month
'''

run_query(query)
```

```python
# No. of recurrin clients from each Lead Source
query = '''
SELECT Lead_Source, SUM(CASE WHEN Ex_Client = 'Yes' THEN 1 ELSE 0 END) AS Recurring_Client
FROM Enquiry_details
GROUP BY 1
ORDER BY 2 DESC
'''

run_query(query)
```

```python
# No. of customized trips
query = '''
SELECT SUM(CASE WHEN Enquiry_Destination LIKE 'BYOG%' OR Enquiry_Destination LIKE '%BYOG' OR Enquiry_Destination LIKE '%BYOG%' THEN 1 END)
AS Cnt_customized_trips
FROM Enquiry_details
'''

run_query(query)
```

```python
# No. of non customized trip
query = '''
SELECT COUNT(*) - SUM(CASE WHEN Enquiry_Destination LIKE 'BYOG%' OR Enquiry_Destination LIKE '%BYOG' OR Enquiry_Destination LIKE '%BYOG%' THEN 1 END)
AS Cnt_non_customized_trips
FROM Enquiry_details
'''

run_query(query)
```

```python
# Days between first Booking and Conversion Date
query = '''
WITH a AS (
SELECT Booking_ID,MIN(SUBSTR(Lead_Date,1,2)) AS first_lead_date, 
Conversion_Date
FROM Enquiry_details
GROUP BY 1
)
SELECT Booking_ID,ABS(SUBSTR(Conversion_Date,1,2) - first_lead_date) AS follow_up_time
FROM a 
ORDER BY 2 DESC
'''

run_query(query)
```

```python
# Leads and RPL analysis per each lead source
query = '''
SELECT Lead_Source, COUNT(*) AS Total_Leads, SUM(Total_Bill) AS Total_Revenue, ROUND(SUM(Total_Bill) / COUNT(*), 2) AS Revenue_Per_Lead
FROM Enquiry_details
WHERE Status = 'Converted'
GROUP BY 1
ORDER BY 4 DESC

'''

run_query(query)
```

```python
# Company's Total ASP
query = '''

SELECT ROUND(SUM(Total_Bill) / SUM(CASE WHEN Status = 'Converted' THEN 1 ELSE 0 END), 2) AS Total_Average_Selling_Price
FROM Enquiry_details
WHERE Status = 'Converted'

'''

run_query(query)
```

```python
# Comparison of Total ASP with Average ASP of each trip
query = '''
WITH Company_ASP AS (
    SELECT ROUND(SUM(Total_Bill) / SUM(CASE WHEN Status = 'Converted' THEN 1 ELSE 0 END), 2) AS Total_Average_Selling_Price
    FROM Enquiry_details
    WHERE Status = 'Converted'
),
Agent_ASP AS (
    SELECT Trip, ROUND(SUM(Total_Bill) / SUM(CASE WHEN Status = 'Converted' THEN 1 ELSE 0 END), 2) AS Trip_Average_Selling_Price
    FROM Enquiry_details
    WHERE Status = 'Converted'
    GROUP BY 1
)
SELECT a.Trip, a.Trip_Average_Selling_Price, c.Total_Average_Selling_Price, ROUND(a.Trip_Average_Selling_Price - c.Total_Average_Selling_Price, 2) AS Difference
FROM  Agent_ASP a, Company_ASP c
ORDER BY 4 DESC
'''

run_query(query)
```

```python
query = '''
WITH Company_ASP AS (
    SELECT ROUND(SUM(Total_Bill) / SUM(CASE WHEN Status = 'Converted' THEN 1 ELSE 0 END), 2) AS Total_Average_Selling_Price
    FROM Enquiry_details
    WHERE Status = 'Converted'
),
Agent_ASP AS (
    SELECT Trip, ROUND(SUM(Total_Bill) / SUM(CASE WHEN Status = 'Converted' THEN 1 ELSE 0 END), 2) AS Trip_Average_Selling_Price
    FROM Enquiry_details
    WHERE Status = 'Converted'
    GROUP BY 1
), Combine_ASP AS (
SELECT a.Trip, a.Trip_Average_Selling_Price, c.Total_Average_Selling_Price, ROUND(a.Trip_Average_Selling_Price - c.Total_Average_Selling_Price, 2) AS Difference
FROM  Agent_ASP a, Company_ASP c
ORDER BY 4 DESC
)
SELECT Combine_ASP.Trip 
FROM Combine_ASP
WHERE Combine_ASP.Difference > 0

'''
```

```python
# Number of repeat customers
query = '''
SELECT Phone, COUNT(*) 
FROM Enquiry_details 
WHERE Status = 'Converted' 
GROUP BY 1 
ORDER BY 2 DESC

'''
run_query(query)
```

```python
#Percentage sales distribution of Agents in the month of July
query = '''
WITH a AS (
SELECT Assigned_To AS Agent, SUBSTR(Conversion_date,4,2) AS Month, COUNT(Phone) AS Agent_Sales
FROM Enquiry_details
WHERE Status = 'Converted' 
GROUP BY 1,2
)
, b AS (
SELECT SUBSTR(Conversion_date,4,2) AS Month, COUNT(Phone) AS Total_Sales
FROM Enquiry_details
WHERE Status = 'Converted'
GROUP BY 1
)
SELECT a.Agent, Agent_Sales*100/Total_Sales AS Contribution
FROM a 
JOIN b
ON a.Month = b.Month
WHERE a.Month = '07'
GROUP BY 1
ORDER BY 2 DESC

'''
run_query(query)
```
